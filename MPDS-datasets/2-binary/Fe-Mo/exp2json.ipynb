```python
import os
import json 
from collections import defaultdict
import copy
```

```python
with open ('Fe-Mo1.exp') as f:
    exp=f.readlines()
    stop=[]
    for i in range(len(exp)):
        exp[i]=exp[i].replace('\n','')
        if 'clip off' in exp[i]:
            stop.append(i)
        if '0.6 ' in exp[i]:
            stop.append(i)
        if 'BLOCK_END' in exp[i]:
            stop.append(i)
    del exp[0]
    
    for i in range(len(stop)-1):
        cont=copy.deepcopy(exp[stop[i]:stop[i+1]-1])
        print(cont)
        if '' in cont:
            cont.remove('')
        if len(cont)==0:
            continue;
        print(cont)
        data_re=defaultdict(list)
        for j in range(len(cont)):
            print(cont[j])
            a=cont[j].split(' ')
            print(a)
            data_re['c1'].append(float(a[0]))
            data_re['c2'].append(float(a[2]))
            
            
            
        print(data_re)
    
        output={}

        values=data_re
    
#    components=input('The components for '+j+':')
        components='FE,MO' #input components
        components=components.split(',')
        output['components']=components
#    phases=input('The phases for '+j+':')
    
        phase_name=['EXP']
        
        phases=phase_name
        output['phases']=phases
        output['broadcast_conditions']='false'
        output['output']='ZPF'
        output['conditions']={}
        output['conditions']['P']=101325
        output['values']=[]
#    output['reference']=input('The reference'+':')
        ref=exp[stop[i+1]-1].split(' ')
        while '' in ref:
            ref.remove('')
        print('ref',ref[3])
        output['reference']=ref[3]#input reference 
#    composition=input('The composition reference for '+j+':')
        composition='MO'#input composition 
        region_data={}
        key_points_number=1
        if key_points_number == 1:
            phase_data=phase_name[0].upper()
            output['conditions']['T']=values['c2']
            for i in values['c1']:
                out=[phase_data,[composition],[i]]
                phases2=copy.deepcopy(phases)
                phases2.remove(phase_data)
                out1=[phases,[composition],['null']]
                output['values'].append([out,out1])
    

        file_name='./ZPF/'+ref[3]+'.json'
        with open(file_name,'w+') as o:
            json.dump(output,o,indent=4)


        out=[]
        n=0
        with open(file_name,'r') as c:
                change=c.readlines()
                work=False
                for i in range(len(change)):
                    len_i=change[i].replace('\n','')
                    len_i=len_i.replace(' ','')
                    if 'values' in change[i]:
                        work=True;
                        len_i='    '+len_i
                    if 'reference' in change[i]:
                        work=False
                    if work: 
                        out[n-1]=out[n-1]+len_i

                    else:
                        out.append(change[i])
                        n=n+1
                print(out)
            name_new='./ZPF/'+'new-'+output['reference']+'.json'
            with open(name_new,'w+') as f:
                for i in out:
                    i=i.replace(']]],',']]],\n')
                    i=i.replace(':[[[',':[\n[[')
                    i=i.replace(']]]],',']]]\n        ],')
                    i=i.replace('[[','             [[')
                    i=i.replace('"null"','null')

                    f.write(i)            





```
